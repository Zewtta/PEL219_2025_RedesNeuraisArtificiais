#### Anotações de aulas:
Reimplemente a rede com 6 camadas para o MNIST usando o Greedy Layer Wise Learning!
- Treine camada a camada.
- Use o atributo “treinable” do Keras para congelar e descongelar camadas.
- Use treinamento supervisionado em cada camada (Adam+Crossentropy)

### Teste 1:
Primeiro teste conforme instruções de aula:
* relu
* softmax
* adam
* GLWT


```python
# load dados do MNIST
from keras.datasets import mnist
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.utils import to_categorical
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# ============================
# 1) Carregar e preparar dados
# ============================
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normaliza os dados (0–1)
x_train = x_train / 255.0
x_test = x_test / 255.0

# One-hot encoding
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)

# ============================
# 2) Definir modelo com 6 camadas
#    (Flatten + 5 densas + saída)
# ============================
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))        # camada 0

model.add(Dense(784, activation='relu'))        # camada 1
model.add(Dense(392, activation='relu'))        # camada 2
model.add(Dense(196, activation='relu'))        # camada 3
model.add(Dense(100, activation='relu'))        # camada 4
model.add(Dense(30,  activation='relu'))        # camada 5
model.add(Dense(10,  activation='softmax'))     # camada 6 (saída)

model.summary()

# ============================
# 3) Treino Greedy Layer-Wise
# ============================

EPOCHS_PER_STAGE = 5
BATCH_SIZE = 32

history_all = []  # se quiser guardar os histories de cada estágio

for i in range(1, len(model.layers)):  # começa depois do Flatten
    print(f"\n=== Treinando a partir da camada {i} / {len(model.layers)-1} ===")

    # congela todas as camadas anteriores
    for layer in model.layers[:i]:
        layer.trainable = False

    # libera as camadas da atual pra frente
    for layer in model.layers[i:]:
        layer.trainable = True

    # recompila o modelo com os flags trainable atualizados
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['categorical_accuracy']
    )

    # treina esse “estágio” da rede
    history = model.fit(
        x_train, y_train,
        validation_split=0.1,
        epochs=EPOCHS_PER_STAGE,
        batch_size=BATCH_SIZE,
        verbose=2
    )
    history_all.append(history)

# ============================
# 4) Fine-tuning final (opcional, mas legal pra tarefa)
# ============================
print("\n=== Fine-tuning com todas as camadas treináveis ===")
for layer in model.layers:
    layer.trainable = True

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['categorical_accuracy']
)

history_final = model.fit(
    x_train, y_train,
    validation_split=0.1,
    epochs=5,
    batch_size=BATCH_SIZE,
    verbose=2
)

# ============================
# 5) Avaliação no teste
# ============================
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"\nTest accuracy (Greedy Layer-Wise + Fine-tuning): {accuracy:.4f}")

# Previsões
y_pred_prob = model.predict(x_test)
y_pred = np.argmax(y_pred_prob, axis=1)
y_true = np.argmax(y_test, axis=1)

# ============================
# 6) Resultados / gráficos
# ============================
model.summary()
print(f'Test accuracy: {accuracy:.4f}')

# (exemplo simples: loss do último treino)
plt.plot(history_final.history['loss'], label='loss treino (fine-tuning)')
plt.plot(history_final.history.get('val_loss', []), label='loss validação')
plt.legend()
plt.xlabel('Época')
plt.ylabel('Loss')
plt.show()

# Matriz de confusão
cm = confusion_matrix(y_true, y_pred)
print("Matriz de Confusão:\n", cm)
print("\nRelatório de Classificação:\n", classification_report(y_true, y_pred))

plt.figure(figsize=(8, 6))
sns.heatmap(
    cm, annot=True, fmt='d', cmap='Blues',
    xticklabels=range(10),
    yticklabels=range(10)
)
plt.title("Matriz de Confusão - MNIST (Greedy Layer-Wise)")
plt.xlabel("Predito")
plt.ylabel("Real")
plt.show()
```
